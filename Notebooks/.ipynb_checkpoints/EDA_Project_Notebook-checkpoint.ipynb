# Global Portfolio Diversification Strategy
---

## Introduction 
In recent years, institutional real estate investing has faced significant headwinds within domestic markets. Aggressive localized inflation, tightening regulatory frameworks, and historic price-to-rent compression have significantly reduced traditional asset yields. For private equity real estate firms and investment funds, remaining solely exposed to a single domestic market risks severe margin erosion and portfolio vulnerability during economic downturns.

To mitigate these risks and secure sustainable long-term revenue growth, funds must transition from local allocation to a dynamic, cross-border diversification strategy. International real estate markets present unique arbitrage opportunities, allowing firms to capitalize on emerging urban centers, favorable currency valuations, and vastly lower entry barriers. By systematically evaluating structural metrics across global listings-such as localized asset inventory, cost per unit area, and construction maturity-investment teams can programmatically uncover undervalued micro-markets that provide superior risk-adjusted returns.

This project applies exploratory data analysis to a global real estate dataset containing over 147,000 listings to establish a rigorous, data-driven framework for international asset acquisition. By transforming raw market parameters into actionable commercial insights, this analysis aims to equip the investment board with the precise geographic and financial targets required to optimize capital deployment and diversify the corporate portfolio.

---

## Problem Statement
Private equity real estate firms face a 15%* stagnation in domestic portfolio yields due to aggressive localized market inflation. Without strategic allocation into alternative assets, investment funds face an anticipated \$2.4M* reduction in annual revenue growth, necessitating the identification of undervalued, fast-growing international micro-markets to diversify property acquisitions before the end of the fiscal year.

Note: *Those are dummy figures inserted to communicate the idea.

## Objectives:

- **Objective 1 (Market Selection):** Which five international countries present the most cost-effective average purchase price per square meter (USD/Area) for contemporary real estate assets?

- **Objective 2 (Asset Optimization):** Within the identified cost-effective countries, how do average purchase prices scale across various bedroom and bathroom configurations, and what constitutes the optimal asset profile for corporate acquisition?

- **Objective 3 (Liquidity and Volume Evaluation):** Which of the target countries contain a sufficient volume of active real estate listings to facilitate large-scale institutional capital deployment without inflating localized market prices?

- **Objective 4 (Capital Allocation Strategy):** Given a designated institutional acquisition budget range of 500,000 to 1,500,000 USD per asset, which specific cities or localized micro-markets within the top-performing countries yield the maximum usable square footage?

### Data Handling: 
__Cleaning, transforming, and combining data__

#### Drop column with no analytical value

```python
# Drop columns that have no analytical value for our business problem
df = df.drop(columns=['image', 'url'], errors='ignore')

# errors='ignore' parameter ensures the code runs without crashing even if one or both of those columns do not exist in the DataFrame
```

#### Fixing data type issues

```python
def clean_area_columns(df, col_name):
    '''extract the first sequence of numbers (including decimals) from a string'''
    if col_name in df.columns:
        # Cast to string, extract matching numeric patterns (digits and optional decimal dots)
        extracted = df[col_name].astype(str).str.extract(r'(\d+[\.]?\d*)')[0]
        # Clean commas if any are used as decimal/thousands separators
        extracted = extracted.str.replace(",", "", regex=False)
        # Convert to float numeric type, coercing unparseable text to NaN
        df[col_name] = pd.to_numeric(extracted, errors='coerce')
    return df
```

```python
# Convert string area columns to numeric values by apply the cleaning function to both area-related string columns
df = clean_area_columns(df, "apartment_total_area")
df = clean_area_columns(df, "apartment_living_area")

# to validate
print(df[["apartment_total_area", "apartment_living_area"]].dtypes)
print(df[["apartment_total_area", "apartment_living_area"]].head(2))
```

#### Handling negative numbers and extreme anomalies using masking

```python
# Replace negative placeholders with NaN to represent true missing values where appropriate
negative_cols = ["building_total_floors", "apartment_floor", "apartment_rooms", "apartment_bedrooms"]
for col in negative_cols:
    df.loc[df[col] < 0, col] = np.nan
```

```python
# Clean up extreme outliers that violate physical possibilities
# 1.Construction Year: Must be between historic limits (year 1800) and current year (2026)
df.loc[(df["building_construction_year"] < 1800) | (df["building_construction_year"] > 2026), "building_construction_year"] = np.nan

# 2. Bedrooms: Restrict to realistic structures (maximum of 15 bedrooms for ultra-luxury/apartments)
df.loc[df["apartment_bedrooms"] > 15, "apartment_bedrooms"] = np.nan

# 3. Rooms: Restrict to realistic room counts (maximum of 30 rooms)
df.loc[df["apartment_rooms"] > 30, "apartment_rooms"] = np.nan

# 4. Rooms: Restrict to realistic bathroom counts (maximum of 10 rooms)
df.loc[df["apartment_bathrooms"] > 10, "apartment_bathrooms"] = np.nan

# 5. Price: Ensure the price is greater than zero
df.loc[df["price_in_USD"] <= 0, "price_in_USD"] = np.nan

# 6. total apartment Area: 
df.loc[df["apartment_total_area"] < 10, "apartment_total_area"] = np.nan
```

#### Handling critical missing values and data type issues

```python
# Drop rows missing crucial parameters for our objectives
# Since Objectives 1 & 4 depend entirely on Price and Total Area, we drop rows missing these two key metrics
df_cleaned = df.dropna(subset=["price_in_USD", "apartment_total_area", "country"])

dropped_rows = len(df) - len(df_cleaned)
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")
print(f"Dropped {dropped_rows} rows due to missing critical evaluation criteria (Price/Area/Country)")
```

```python
# convert discrete count and year columns to Nullable Integers (Int64) (capital I)
integer_targets = ["building_construction_year", "building_total_floors", "apartment_floor",
                   "apartment_rooms", "apartment_bedrooms", "apartment_bathrooms"]

for col in integer_targets:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].astype("Int64")

# Verify the types updated
print(df_cleaned[integer_targets].dtypes)
```

#### Feature Engineering

```python
# Calculating Price per Square Meter (price_per_sqm)
df_cleaned["price_per_sqm"] = df_cleaned["price_in_USD"] / df_cleaned["apartment_total_area"]

# Calculating the age of the building as of current year (2026)
df_cleaned["building_age"] = 2026 - df["building_construction_year"]

# Preview the engineered features
df_cleaned[["price_in_USD", "apartment_total_area", "price_per_sqm", "building_age"]].head()
```

```python
# Convert the new column "building_age" to Int64
df_cleaned["building_age"] = df_cleaned["building_age"].astype("Int64")
```

```python
# To validate the changes made
df_cleaned.info()
```

### Analysis: 
__Answering the objectives through data analysis__



#### Objective 1: Market Selection (Top 5 Cost-Effective Countries)

```python
# 1. Applying a boolean mask to isolate contemporary properties (built between 2011 and 2026) "last 15 years"
contemporary_mask = (df_cleaned["building_construction_year"] >= 2011) & (df_cleaned["building_construction_year"] >= 2026)
df_contemporary = df_cleaned[contemporary_mask]

# 2. Grouping by country, calculate mean price per sqm, and find the volume of properties
country_metrics = df_contemporary.groupby("country").agg(avg_price_per_sqm=('price_per_sqm', 'mean'), listing_count=('price_per_sqm', 'count')).reset_index()

# 3. Sorting by price per square meter ascending and extract the top 5 most cost-effective countries
# Note: We filter for countries with at least 100 listings to avoid thin-data outliers --> modified to 45 for some reasons related to the data itself
top_5_countries_df = country_metrics[country_metrics["listing_count"] >= 45].sort_values(by="avg_price_per_sqm", ascending=True).head(5)

print("Top 5 Most Cost-Effective Countries (Built 2011-2026)")
print(top_5_countries_df)

# Storing the names of the top 5 countries for downstream multi-criteria masking
top_5_countries = top_5_countries_df["country"].tolist()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_countries_df, x='avg_price_per_sqm', y='country')


plt.title('Top 5 Most Cost-Effective Countries by Avg Price per Sqm (Built 2011-2026)', fontsize=14, weight='bold')
plt.xlabel('Avg Price per Square Meter (USD)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tick_params(axis='both', labelsize=11)
plt.tight_layout()
plt.grid(False)
plt.show()
```

#### Objective 2: Asset Optimization (Top 5 Most Cost-Effective Asset Configuration)

```python
# 1. Filter the cleaned dataset to include only the selected target countries
market_optimization_mask = df_cleaned['country'].isin(top_5_countries)
df_target_markets = df_cleaned[market_optimization_mask]

# 2. Aggregate price and size metrics by bedroom and bathroom configurations
# We filter out rare combinations (e.g., configurations with fewer than 10 listings) to keep trends robust
assets_profile_metrics = df_target_markets.groupby(['apartment_bedrooms', 'apartment_bathrooms']).agg(
    avg_price_in_USD = ('price_in_USD', 'mean'),
    avg_price_per_sqm = ('price_per_sqm', 'mean'),
    avg_total_area = ('apartment_total_area', 'mean'),
    listing_count = ('price_in_USD', 'count')
).reset_index()

# 2. Filter for statistically significant asset volume
df_optimized_profiles = assets_profile_metrics[assets_profile_metrics['listing_count'] >= 10]

# 3. Sort by price per square meter to uncover the most optimized layout profiles
df_optimized_profiles = df_optimized_profiles.sort_values(by= 'avg_price_per_sqm', ascending=True)

# Preview the Top 5 Most Cost-Effective Asset Configurations
df_optimized_profiles.head()
```

```python
# Technical Report Visualization for Objective 2
# Creating a pivot table to construct a clear, scannable heatmap of price per sqm across layout profiles
pivot_heatmap = df_optimized_profiles.pivot(
    index='apartment_bedrooms',
    columns='apartment_bathrooms',
    values='avg_price_per_sqm'
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_heatmap, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Avg Price per Sqm (USD)'})

plt.title('Avg Price per Sqm by Bedroom & Bathroom Layout', fontsize=14, weight='bold')
plt.xlabel('Number of Bathrooms', fontsize=12)
plt.ylabel('Number of Bedrooms', fontsize=12)
plt.tick_params(axis='both', labelsize=11)
plt.tight_layout()
plt.show()
```

#### Objective 3: Liquidity and Volume Evaluation (Total Available Listings by Target Country)

```python
# Calculate market volume and liquidity metrics across our target countries
market_liquidity = df_target_markets.groupby('country').agg(
    total_listings = ('price_in_USD', 'count'),
    median_price_USD = ('price_in_USD', 'median'),
    avg_building_age = ('building_age', 'mean')
).reset_index().sort_values(by="total_listings", ascending=False)

# Market Liquidity and Supply Volume Analysis
market_liquidity
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=market_liquidity, x='total_listings', y='country')


plt.title('Total Available Listings by Target Country', fontsize=14, weight='bold')
plt.xlabel('Total Active Property Listings (Volume)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tick_params(axis='both', labelsize=11)
plt.tight_layout()
plt.grid(False)
plt.show()
```

#### Objective 4: Capital Allocation Strategy (Top 10 Micro-Markets by Space Yield)

```python
# Apply multi-criteria boolean masking for the institutional budget range
budget_mask = (df_target_markets['price_in_USD'] >= 500000) & (df_target_markets['price_in_USD'] <= 1500000)
df_budget_isolated = df_target_markets[budget_mask]

# Group by location/city and calculate average usable area and median price
city_allocation = df_budget_isolated.groupby(['country', 'location']).agg(
    avg_total_area = ('apartment_total_area', 'mean'),
    median_price_USD = ('price_in_USD', 'median'),
    listing_count = ('price_in_USD', 'count')    
).reset_index()

# Filter for micro-markets with a baseline volume of listings to ensure stable entry
city_allocation = city_allocation[city_allocation['listing_count'] >= 5]

# Sort by average total area descending to uncover where capital buys the most space
top_10_cities = city_allocation.sort_values(by='avg_total_area', ascending=False).head(10)

# Top 10 Micro-Markets by Space Yield
top_10_cities
```

```python
# Creating a combined location string for clear chart formatting

plt.figure(figsize=(12,6))
sns.barplot(data=top_10_cities, x='avg_total_area', y='location')


plt.title('Top 10 Micro-Markets by Average Property Space (\$500K-\$1.5M Budget)', fontsize=14, weight='bold')
plt.xlabel('Average Total Area (Square Meters)', fontsize=12)
plt.ylabel('Location (Country)', fontsize=12)
plt.tick_params(axis='both', labelsize=11)
plt.tight_layout()
plt.grid(False)
plt.show()
```

## Summary
__Summarizing the key insights from the analysis__

- **Geographic Price Disparity (Objective 1):** Georgia, Russia, Thailand, and Northern Cyprus emerge as exceptionally cost-effective entry points for contemporary assets, all exhibiting baseline valuations under 5,000 USD per square meter. Conversely, the United Arab Emirates (UAE) represents an extreme premium market, ballooning to an average exceeding 80,000 USD per square meter.

- **Asset Profile Layout Scaling (Objective 2):** Real estate valuations scale efficiently within standard configurations (1 to 4 bedrooms), where average costs range between a highly competitive 1937.6 and 8,791.5 USD per square meter. However, a major structural pricing pivot occurs at highly dense layouts, marked by extreme institutional pricing premiums for 5-bedroom, 6-bathroom configurations which exceed 12137.3 USD per square meter.

- **Market Scale and Liquidity (Objective 3):** Russia dominates the analyzed footprint with a massive volume of over 17,500 active property listings, signaling an incredibly deep, high-liquidity market. Georgia and the UAE provide healthy moderate capacity (roughly 2,500 to 3,200 listings), while Northern Cyprus displays restricted inventory under 1,000 listings, presenting a higher risk of localized price inflation if large corporate capital is deployed rapidly.

- **Space Optimization Under Allocation Targets (Objective 4):** Within the corporate budget window of 500K to 1.5M USD, administrative micro markets surrounding Moscow, Russia (Pervomayskoe and Marushkinskoe) yield the maximum spatial capacity, offering massive properties spanning between 400 and 550 square meters. For tropical or alternative diversification, Ban Kata in Phuket, Thailand and Batumi in Georgia offer robust mid-tier space configurations spanning approximately 300 to 335 square meters.

## Recommendations/Conclusion

- **Execute Core Portfolio Capital Deployment in Russia and Georgia:** Allocate the primary portion of the diversification fund to the Russian and Georgian contemporary micro-markets. These regions successfully combine exceptional cost efficiencies (under$3,000/sqm) with strong inventory volume to ensure scalable, smooth institutional acquisition programs.

- **Target Mid-Tier 2-to-3 Bedroom Configurations:** Prioritize the acquisition of real estate assets containing 2 to 3 bedrooms and 1 to 2 bathrooms. The analysis proves these structural profiles occupy the financial "sweet spot," allowing the firm to maximize spatial footprint while keeping asset valuations constrained within the lowest price-per-square-meter bands.

- **Impose a Strict Prohibition on High-Density Luxury Layouts:** Issue an immediate corporate mandate restricting the purchase of properties with 5 or more bedrooms or 4 or more bathrooms. The data confirms these assets carry heavily inflated luxury premiums (surpassing $12K/sqm) that would severely impair the portfolio's aggregate return on investment (ROI).

- **Deploy Targeted Capital to Specific High-Yield Micro-Locations:** Instruct acquisition teams to focus the targeted 500K-1.5M USD budget on specific high-yielding cities: Pervomayskoe (Russia) for maximum corporate asset scale, Batumi (Georgia) for emerging urban exposure, and Ban Kata, Phuket (Thailand) to secure high-footprint, premium geographic alternatives.
